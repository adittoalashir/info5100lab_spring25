## Scatter Plot

### Bivariate
Visual representation of the relationship between two variables in a dataset.

````python
plt.grid(True)
plt.scatter(iris_dataframe["age"], iris_dataframe["education-num"])
plt.title("Age vs Eductaion-num")
plt.xlabel("Age")
plt.ylabel("Education-num")
````

### Multivariate

## Histograms

````python
sns.histplot(data=iris_dataframe, x="gender")
plt.title("Gender distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
````

````python
sns.histplot(data=iris_dataframe, x="race")
plt.title("Race Distribution")
plt.grid(axis="y")
plt.xlabel("Race")
plt.ylabel("Count")
plt.yticks(range(0, 30, 2))
````

## Kernel Density Plot (KDE)

KDE plot helps to visualize the distribution of data in a dataset



````python
# Create a KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=iris_dataframe, x="age", hue="occupation", fill=True, common_norm=False, palette="crest", alpha=0.2)

# Add labels and title
plt.title("Kernel Density Plot of Occupation by age")
plt.xlabel("age")
plt.ylabel("occupation")
plt.show()
````

## Violin Plot (Box Plot + KDE)

A statistical graphic that shows the distribution of numerical data by combining a box plot and a kernel density plot.

````python
sns.violinplot(x='occupation', y='hours-per-week', data=iris_dataframe)
plt.title('Violin Plot of occupation and hours-per-week')
plt.show()
````
